```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
import random
import time
import bs4
import pandas as pd
from selenium.webdriver.common.action_chains import ActionChains
```

```python
def open_website(URL='https://www.amazon.com/RockBirds-Flashlights-Bright-Aluminum-Flashlight/product-reviews/B00X61AJYM'):
    driver = webdriver.Firefox(executable_path='/Users/heli/Desktop/geckodriver')
    driver.get(URL)
    return driver
```

```python
driver = open_website('https://www.amazon.com/RockBirds-Flashlights-Bright-Aluminum-Flashlight/product-reviews/B00X61AJYM')
```

```python
driver.execute_script("window.scrollBy(0,800)")
```

```python
recent_name=driver.find_element_by_xpath('//*[@id="a-autoid-4-announce"]')
recent_name.click()
```

```python
recent=driver.find_element_by_xpath('//*[@id="sort-order-dropdown_1"]')
recent.click()
```

```python

reviewers_sort=driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[1]/div[2]/div/div[1]/div[2]/div[2]/div[2]/a[1]/span/span/span/span/span')
reviewers_sort.click()
```

```python
reviewers_buy=driver.find_element_by_xpath('//*[@id="reviewer-type-dropdown_1"]')
reviewers_buy.click()
```

```python
Name = []
Date = []
Rating=[]
Reviews=[]
Review_title=[]

wait=WebDriverWait(driver,10)
page_number=1

while page_number <= 80:
    delay = wait.until(EC.presence_of_element_located((By.ID,'cm_cr-review_list')))
    time.sleep(2)
    
    element = driver.find_element_by_id('cm_cr-review_list')
    data_html = element.get_attribute('innerHTML')
    soup=bs4.BeautifulSoup(data_html,'html5lib')


    name_element=[i.text for i in soup.find_all('a',attrs={'class': 'a-size-base a-link-normal author'})]

    date_element = [i.text for i in soup.find_all('span',{'class':'a-size-base a-color-secondary review-date'})]

    rating_element = [i.text.replace(' out of 5 stars','') for i in soup.find_all('i',{'data-hook':'review-star-rating'})]

    reviews_element = [i.text for i in soup.find_all('span',{'data-hook':'review-body'})]
    review_title_element = [i.text for i in soup.find_all('a',{'data-hook':'review-title'})]
    

    for i in range(len(name_element)):
        Name.append(name_element[i])     
        Date.append(date_element[i])
        Rating.append(rating_element[i])
        Reviews.append(reviews_element[i])
        Review_title.append( review_title_element[i])
        
  
    time.sleep(2)
    next_button = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'a-last'))) 
    
    next_button.click()
    
    page_number+=1
    
    
```

```python
total_data=[]
for i in range(len(Name)):
    total_data.append([Name[i],Date[i],Rating[i],Reviews[i],Review_title[i]])

        
```

```python
df=pd.DataFrame(total_data)
```

```python
for i in range(793,800):
    df.drop([i],inplace=True)
```
